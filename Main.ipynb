```python
import pandas as pd
import numpy as np
df = pd.read_csv('/content/drive/MyDrive/MILLENIUM PRO-1/WH_data.csv')

# Separate categorical and numerical columns
categorical_columns = df.select_dtypes(include=['object', 'category']).columns.tolist()
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

print("Categorical columns: ", categorical_columns)
print("Numerical columns: ", numerical_columns)

# Basic analysis on categorical columns
for col in categorical_columns:
    print(f"\nColumn: {col}")
    print(df[col].value_counts())

# Basic analysis on numerical columns
df[numerical_columns].describe()
```

```python
#Find the total count of countries?
num_countries = df['Country'].nunique()
print(f"Total number of unique countries: {num_countries}")
```

```python
#Find first 10 rows and last 10 rows from data.
first_10_rows = df.head(10)
print("First 10 rows of the DataFrame:")
print(first_10_rows)
last_10_rows = df.tail(10)
print("\nLast 10 rows of the DataFrame:")
print(last_10_rows)
```

```python
#Find the total count of countries from each region?
countries_per_region = df.groupby('Region')['Country'].nunique()

print("Number of unique countries in each region:")
print(countries_per_region)
```

```python
#Give me the list of regions
regions = df['Region'].unique()
print("List of unique regions:")
for region in regions:
    print(region)
```

```python
average_gdp_per_region = df.groupby('Region')['Economy (GDP per Capita)'].mean()
```

```python
print(average_gdp_per_region)
```

```python
max_health = df['Health (Life Expectancy)'].max()
min_health = df['Health (Life Expectancy)'].min()
average_health = df['Health (Life Expectancy)'].mean()

print(f'The maximum Health(Life expectancy) is {max_health}')
print(f'The minimum Health(Life expectancy) is {min_health}')
print(f'The average Health(Life expectancy) is {average_health}')
```

```python
max_trust = df['Trust (Government Corruption)'].max()
min_trust = df['Trust (Government Corruption)'].min()
average_trust = df['Trust (Government Corruption)'].mean()

print(f'The maximum Trust (Government Corruption) is {max_trust}')
print(f'The minimum Trust (Government Corruption) is {min_trust}')
print(f'The average Trust (Government Corruption) is {average_trust}')
```

```python
year = 2017
average_health = df[df['Year'] == year]['Health (Life Expectancy)'].mean()

print(f'The average Health(Life expectancy) in the year {year} is {average_health}')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

average_health_per_year = df.groupby('Year')['Health (Life Expectancy)'].mean()
plt.figure(figsize=(10, 6))
plt.plot(average_health_per_year.index, average_health_per_year.values)
plt.xlabel('Year')
plt.ylabel('Average Health(Life expectancy)')
plt.title('Average Health(Life expectancy) Over the Years')
plt.grid(True)
plt.show()
```

```python
average_health_per_year = df.groupby('Year')['Health (Life Expectancy)'].mean()
plt.figure(figsize=(10, 6))
plt.bar(average_health_per_year.index, average_health_per_year.values)
plt.xlabel('Year')
plt.ylabel('Average Health(Life expectancy)')
plt.title('Average Health(Life expectancy) Over the Years')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
summary = df.groupby(['Region', 'Country', 'Year'])['Freedom'].describe()

print(summary)
plt.figure(figsize=(10, 6))
sns.boxplot(x='Region', y='Freedom', data=df)
plt.xticks(rotation=90)
plt.title('Distribution of Freedom by Region')
plt.show()
average_freedom_per_year = df.groupby('Year')['Freedom'].mean()
plt.figure(figsize=(10, 6))
plt.plot(average_freedom_per_year.index, average_freedom_per_year.values)
plt.xlabel('Year')
plt.ylabel('Average Freedom')
plt.title('Average Freedom Over the Years')
plt.grid(True)
plt.show()
```

1.Line Chart: This is useful for visualizing a trend in data over intervals of time - a time series.
2.Bar Chart: This is useful when you want to compare a single category of data between individual sub-items.
3.Histogram: This is useful for visualizing the underlying frequency distribution (shape) of a set of continuous data.
4.Scatter Plot: This is useful for visualizing relationships between two numerical variables where one variable is independent.
5.Pie Chart: This is useful when you want to visualize a categorical dataset and see the proportion of each category in the data.
6.Box Plot: This is useful for visualizing the statistical summary of a data set including minimum, first quartile, median, third quartile, and maximum.
7.Heatmap: This is useful for visualizing correlation between multiple variables in a dataset.
8.Pair Plot: This is useful for visualizing pairwise relationships in a dataset.
9.Area Chart: This is useful for visualizing quantitative data graphically.
